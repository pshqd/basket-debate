```python
import pandas as pd

path = db_path  # если это путь к .tsv/.csv выгрузке, а не к products.db

df_head = pd.read_csv(
    path,
    sep="\t",
    nrows=50,
    low_memory=False
)
df_head.columns[:30], df_head.head(3)
```

```python
usecols = ['code', 'url', 'creator', 'created_t', 'created_datetime',
        'last_modified_t', 'last_modified_datetime', 'last_modified_by',
        'last_updated_t', 'last_updated_datetime', 'product_name',
        'abbreviated_product_name', 'generic_name', 'quantity', 'packaging',
        'packaging_tags', 'packaging_en', 'packaging_text', 'brands',
        'brands_tags', 'brands_en', 'categories', 'categories_tags',
        'categories_en', 'origins', 'origins_tags', 'origins_en',
        'manufacturing_places', 'manufacturing_places_tags', 'labels']

chunks = pd.read_csv(
    path,
    sep="\t",
    usecols=lambda c: c in usecols,
    chunksize=500,
    low_memory=False,
    encoding_errors="replace",
)

for chunk in chunks:
    print(chunk.info())
    break
```

```python
for chunk in chunks:
    print(chunk.describe().transpose())
    break
```

```python
for chunk in chunks:
    print(chunk.describe(include='object').transpose())
    break
```
